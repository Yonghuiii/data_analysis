## 2. 구_행정코드 불러오기

```python
구동코드 = pd.read_excel('C:\\Users\kwaks02\유통\GS리테일.xlsx', sheet_name='참고)구_행정동코드',
                    skiprows=[0],
                    skip_cols=[0,1])
```

```python
구코드 = 구동코드[['구코드','구명']]
동코드 = 구동코드[['행정동코드', '행정동명']]
```

## 3. 구명을 종합테이블에 붙이기

```python
구명 =[]
for i in range(len(GS)):
    if GS.iloc[i]['구_CD'] == 1111:
        구명.append('종로구')
    else:
        구명.append('노원구')
```

```python
구명 = pd.DataFrame(구명, columns = ['구명'])
GS = pd.concat([GS, 구명], axis=1)
```

## 4. 행정동명을 종합테이블에 붙이기

```python
행정동명 =[]
for i in range(len(GS)):
    for j in range(len(동코드)):
        if GS.iloc[i]['동_CD'] == 동코드.iloc[j]['행정동코드']:
            행정동명.append(동코드.iloc[j]['행정동명'])
```

```python
행정동명 = pd.DataFrame(행정동명, columns = ['행정동명'])
GS = pd.concat([GS, 행정동명], axis=1)
```

```python
GS = GS[['DT', '구_CD', '구명', '동_CD', '행정동명', '매출지수', 
   '식사', '간식', '마실거리', '홈_리빙', '헬스_뷰티', '취미_여가', '사회활동', '임신/육아']]
```

```python
GS.head()
```

## 5. EDA - 종로구와 노원구 매출지수 차이

```python
구별매출지수 = GS.pivot_table(index='DT',
              columns = '구명',
              aggfunc = 'mean',
              values = '매출지수')
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'].insert(0, 'Malgun Gothic') 

구별매출지수.plot(figsize = (20,10), color=['orange','green'])
plt.legend(fontsize=20)
plt.show()

# 종로구가 노원구보다 매출이 더 높은 것을 알 수 있다.
```

### - 구별매출최대값이 기록된 날짜구하기

```python
구별매출최대값 = GS.pivot_table(index='DT',
              columns='구명',
              aggfunc='max',
              values='매출지수')
구별매출최대값.head()
```

```python
구별매출최대값[구별매출최대값.종로구 == 구별매출최대값.종로구.max()]['종로구']
```

```python
구별매출최대값[구별매출최대값.노원구 == 구별매출최대값.노원구.max()]['노원구']
```
