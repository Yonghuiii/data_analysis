## # 지하철 호선별 역별 승하차인원 데이터 합치기(18년4월 ~ 19년3월) 

```python
# 18년도 4월 ~ 19년도 3월 지하철 호선별 역별 승하차인원 데이터 합치기

months_18_1 = range(4,10)

data = []

for month in months_18_1:
    path = 'C:\Python\JupyterNotebook\[공모전]빅콘\DATASET(미세먼지)\data\지하철\CARD_SUBWAY_MONTH_20180%d.csv' % month
    frame = pd.read_csv(path, index_col=False)
    
    frame['월'] = month
    data.append(frame)

지하철유동인구_18_1 = pd.concat(data, ignore_index=True)


months_18_2 = range(10,13)

for month in months_18_2:
    path = 'C:\Python\JupyterNotebook\[공모전]빅콘\DATASET(미세먼지)\data\지하철\CARD_SUBWAY_MONTH_2018%d.csv' % month
    frame = pd.read_csv(path, index_col=False)
    
    frame['월'] = month
    data.append(frame)

지하철유동인구_18_2 = pd.concat(data, ignore_index=True)
지하철유동인구_18_2.head()
#위의 결과에 19년도 성연령유인구 파일 추가하기

months_19 = range(1,4)

for month in months_19:
    path = 'C:\Python\JupyterNotebook\[공모전]빅콘\DATASET(미세먼지)\data\지하철\CARD_SUBWAY_MONTH_20190%d.csv' % month
    frame = pd.read_csv(path, index_col=False)
   
    frame['월'] = month
    data.append(frame)

지하철유동인구 = pd.concat(data, ignore_index=True)

지하철유동인구
```

```python
from __future__ import print_function
import os
from PIL import Image

files = [
      '경복궁역 승차총승객수(사직동).jpg','광화문역 승차총승객수(사직동).jpg','종각역 승차총승객수(종로1,2,3,4가동).jpg','안국역 승차총승객수(삼청동).jpg',
    '종로3가역 승차총승객수(종로1,2,3,4가동).jpg','종로5가역 승차총승객수(종로1,2,3,4가동_5,6가동).jpg','혜화역 승차총승객수(이화동).jpg','동묘앞역 승차총승객수(숭인1동).jpg']

result = Image.new("RGB", (1200, 800))

for index, file in enumerate(files):
    path = os.path.expanduser(file)
    img = Image.open(path)
    img.thumbnail((250, 350), Image.ANTIALIAS)
    x = index // 2 * 300
    y = index % 2 * 400
    w, h = img.size
    result.paste(img, (x, y, x + w, y + h))

result.save(os.path.expanduser('종로구_지하철유동인구(전체).jpg'))
```
