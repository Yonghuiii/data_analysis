```python
import os
import numpy as np
import pandas as pd
```

```python
filelist = [i for i in list(os.walk('/Users/Neun/Desktop/bigcon/innovation/온습도_평균_표준편차'))[0][2] if 'csv' in i]
donglist = [i.split('.')[0] for i in filelist]
file_dong = list(zip(filelist, donglist))
file_dong
```

```python
# 온습도 데이터 적재
temp_humi = {}
for f, d in file_dong:
    temp_humi[d] = pd.read_csv('/Users/Neun/Desktop/bigcon/innovation/온습도_평균_표준편차/' + f)
    if 'Unnamed: 0' in temp_humi[d].columns:
        temp_humi[d].drop('Unnamed: 0', axis = 1, inplace = True)

```

```python
# 데이터 컬럼명 통일해주고

for dong in temp_humi.keys():
    if temp_humi[dong].columns[0] == 'temp':
            temp_humi[dong].columns = ['date', '습도_mean', '온도_mean', '습도_std', '온도_std']

for dong in temp_humi.keys():
    colnames = []
    for col in temp_humi[dong]:
        if col == 'temp_mean':
            colnames.append('온도_mean')
        elif col == 'temp_std':
            colnames.append('온도_std')

        elif col == 'humi_mean':
            colnames.append('습도_mean')
        elif col == 'humi_std':
            colnames.append('습도_std')

        else:
            colnames.append(col)
    temp_humi[dong].columns = colnames
            
for dong in temp_humi.keys():
    print(temp_humi[dong].columns)
```

```python
for dong in temp_humi.keys():
    temp_humi[dong]['date'] = pd.to_datetime(temp_humi[dong]['date'], format = '%Y-%m-%d')
    temp_humi[dong].set_index(temp_humi[dong]['date'], inplace = True)
    temp_humi[dong].drop('date', axis = 1, inplace = True)
    
```

```python
temp_humi['이화동'].head()
```

```python
for dong in temp_humi.keys():
    print(dong)
    print(len(temp_humi[dong].index.month.value_counts().sort_index()))
    print(temp_humi[dong].index.month.value_counts().sort_index())
    print()
```

```python
# 월별 평균 만들고
temp_humi_mean = {}
for dong in set(temp_humi.keys()).difference(set(['상계1동', '월계3동', '중계2,3동'])):
    temp_humi_mean[dong] = \
    temp_humi[dong].pivot_table(index = temp_humi[dong].index.month, values = ['습도_mean', '온도_mean', '습도_std', '온도_std'], aggfunc = 'mean').reset_index()
    temp_humi_mean[dong].columns = ['월', '습도_mean_월평균', '습도_std_월평균', '온도_mean_월평균', '온도_std_월평균']
```

```python
temp_humi_mean['이화동'].head()
```

```python
date_all = pd.DataFrame(pd.date_range(start = '2018-04-01', end = '2019-03-31', freq = 'D'), columns = ['date'])
```

```python
# 일단 모든 동에 대해서 모든 일자를 다 만들어줌
temp_humi_all = {}
for dong in temp_humi.keys():
    temp_humi_all[dong] = pd.merge(temp_humi[dong], date_all, on = 'date' , how = 'outer')
    temp_humi_all[dong]['월'] = [i.month for i in temp_humi_all[dong]['date']]  # 이 때 월도 제대로 만들어주고
    
```

'date', '온도_mean', '온도_std', '습도_mean', '습도_std'  
이 순서대로 만들어놓음!

```python
for dong in temp_humi_mean.keys():
    temp_humi_all[dong] = pd.merge(temp_humi_all[dong], temp_humi_mean[dong], on = '월', how = 'left')
```

```python
for dong in temp_humi_mean.keys():
    temp_humi_all[dong] = temp_humi_all[dong].reindex(sorted(temp_humi_all[dong].columns), axis=1)
```

```python
for dong in temp_humi_all.keys():
    print(list(temp_humi_all[dong].columns))
```

```python
for dong in temp_humi_mean.keys():
    temp_mean, humi_mean, temp_std, humi_std = [], [], [], []
    for row in temp_humi_all[dong].values:
# 습도 평균
        if np.isnan(row[1]):
            humi_mean.append(row[2])
        else:
            humi_mean.append(row[1])
# 습도 표준편차            
        if np.isnan(row[3]):
            humi_std.append(row[4])
        else:
            humi_std.append(row[3])
# 온도 평균
        if np.isnan(row[5]):
            temp_mean.append(row[6])
        else:
            temp_mean.append(row[5])
# 온도 표준편차            
        if np.isnan(row[7]):
            temp_std.append(row[8])
        else:
            temp_std.append(row[7])
    
    temp_humi_all[dong]['온도_mean_new'] = temp_mean
    temp_humi_all[dong]['온도_std_new'] = temp_std
    temp_humi_all[dong]['습도_mean_new'] = humi_mean
    temp_humi_all[dong]['습도_std_new'] = humi_std
```

```python
temp_humi_all['상계10동'][temp_humi_all['상계10동']['온도_mean'].isna()]
```

```python
월계3_replace = pd.merge(temp_humi_all['공릉1동'][['date', '온도_mean_new', '온도_std_new', '습도_mean_new', '습도_std_new']], \
         temp_humi_all['월계1동'][['date', '온도_mean_new', '온도_std_new', '습도_mean_new', '습도_std_new']], on = 'date')

월계3_replace['온도_mean_평균'] = 월계3_replace[['온도_mean_new_x', '온도_mean_new_y']].mean(axis= 1)
월계3_replace['온도_std_평균'] = 월계3_replace[['온도_std_new_x', '온도_std_new_y']].mean(axis= 1)

월계3_replace['습도_mean_평균'] = 월계3_replace[['습도_mean_new_x', '습도_mean_new_y']].mean(axis= 1)
월계3_replace['습도_std_평균'] = 월계3_replace[['습도_std_new_x', '습도_std_new_y']].mean(axis= 1)

월계3_replace.drop(['온도_mean_new_x', '온도_std_new_x', '습도_mean_new_x',
       '습도_std_new_x', '온도_mean_new_y', '온도_std_new_y', '습도_mean_new_y',
       '습도_std_new_y'], axis = 1, inplace = True)
월계3_replace.head()
```

```python
temp_humi_all['월계3동'] = pd.merge(temp_humi_all['월계3동'], 월계3_replace, on = 'date', how = 'left')
temp_humi_all['월계3동'] = temp_humi_all['월계3동'].reindex(sorted(temp_humi_all['월계3동'].columns), axis=1)
temp_humi_all['월계3동'].head()


temp_mean, humi_mean, temp_std, humi_std = [], [], [], []
for row in temp_humi_all['월계3동'].values:
# 습도 평균
        if np.isnan(row[1]):
            humi_mean.append(row[2])
        else:
            humi_mean.append(row[1])
# 습도 표준편차            
        if np.isnan(row[3]):
            humi_std.append(row[4])
        else:
            humi_std.append(row[3])
# 온도 평균
        if np.isnan(row[5]):
            temp_mean.append(row[6])
        else:
            temp_mean.append(row[5])
# 온도 표준편차            
        if np.isnan(row[7]):
            temp_std.append(row[8])
        else:
            temp_std.append(row[7])
    

temp_humi_all['월계3동']['온도_mean_new'] = temp_mean
temp_humi_all['월계3동']['온도_std_new'] = temp_std
temp_humi_all['월계3동']['습도_mean_new'] = humi_mean
temp_humi_all['월계3동']['습도_std_new'] = humi_std

temp_humi_all['월계3동'].head()
```

```python
중계23_replace = pd.merge(temp_humi_all['상계67동'][['date', '온도_mean_new', '온도_std_new', '습도_mean_new', '습도_std_new']], \
         temp_humi_all['하계1동'][['date', '온도_mean_new', '온도_std_new', '습도_mean_new', '습도_std_new']], on = 'date')

중계23_replace['온도_mean_평균'] = 중계23_replace[['온도_mean_new_x', '온도_mean_new_y']].mean(axis= 1)
중계23_replace['온도_std_평균'] = 중계23_replace[['온도_std_new_x', '온도_std_new_y']].mean(axis= 1)

중계23_replace['습도_mean_평균'] = 중계23_replace[['습도_mean_new_x', '습도_mean_new_y']].mean(axis= 1)
중계23_replace['습도_std_평균'] = 중계23_replace[['습도_std_new_x', '습도_std_new_y']].mean(axis= 1)

중계23_replace.drop(['온도_mean_new_x', '온도_std_new_x', '습도_mean_new_x',
       '습도_std_new_x', '온도_mean_new_y', '온도_std_new_y', '습도_mean_new_y',
       '습도_std_new_y'], axis = 1, inplace = True)
중계23_replace.head()
```
