### Serial명 -> 행정동명 변경 함수(노원구, 종로구도 똑같이 관측소명과 행정동명 매칭해주면 됨)

```python

def 관측소_to_행정동_노원(serial):

    if serial=='V10O1610102' or serial=='V10O1610610':
        serial = '하계1동'
    elif serial=='V10O1610616':
        serial = '상계2동'
    elif serial=='V10O1610200' or serial=='V10O1610293' or serial=='V10O1610312' \
        or serial=='V10O1610351' or serial=='V10O1610356' or serial=='V10O1610376' \
        or serial=='V10O1611100' or serial=='V10O1611102' or serial=='V10O1611104'\
        or serial=='V10O1611150':
        serial = '상계3.4동'
    elif serial=='V10O1610297':
        serial = '상계5동'
    elif serial=='V10O1610629' or serial=='V10O1610630' or serial=='V10O1610642' \
        or serial=='V10O1611652':
        serial = '공릉2동'
    elif serial=='V10O1610643' or serial=='V10O1611229':
        serial = '월계1동'
    elif serial=='V10O1611097' or serial=='V10O1612126':
        serial = '중계본동'
    else:
        serial = '없는 관측소입니다'
    
    return serial
```

```python

def 관측소_to_행정동_종로(serial):

    if serial=='V10O1610252':
        serial = '교남동'
    elif serial=='V10O1610546' or 'V10O1610540' or 'V10O1610542' or 'V10O1610543' or 'V10O1610544' or 'V10O1610545' or 'V10O1610567' \
        or serial== 'V10O1612113' or 'V10O1611145' or 'V10O1611750' or 'V10O1611684' or 'V10O1611173':
        serial = '종로 1,2,3,4가동'
    elif serial=='V01o1610468' or 'V10O1611172':
        serial = '사직동'
    elif serial=='V10O1611289':
        serial = '천연동'
    elif serial=='V10O1611634':
        serial = '가회동'
    elif serial=='V10O1611887':
        serial = '명동'
    elif serial=='V10O1611639':
        serial = '종로5,6가동'
    elif serial=='V10O1611658':
        serial = '이화동'   
    elif serial=='V10O1611151' or 'V10O1611220' or '':
        serial = '창신1동' 
    elif serial=='V10O1611623' or 'V10O1611645' or '':
        serial = '혜화동'
    elif serial=='V10O1611170':
        serial = '부암동'
    elif serial=='V10O1612106':
        serial = '숭인2동'  
    elif serial=='V10O1611251':
        serial = '창신3동'
    elif serial=='V10O1611258':
        serial = '평창동'
    elif serial=='V10O1611255':
        serial = '청운효자동'
    elif serial=='V10O1611698' or 'V10O1611722':
        serial = '가회동'           
    else:
        serial = '없는 관측소입니다'
    
    return serial
```

### 데이터 적재 및 전처리 함수(검토 후 전처리 부분은 더욱 고민할 필요있음)
### 데이터 전처리 구현상황 : 불필요 컬럼 제거 및 결측값 0으로 채우기, 이상값(-999,-9999) 제거

```python
def 데이터전처리_노원(file):
    serial = file.split('/')
    serial_name = serial[-1]
    serial = serial_name.split('.')
    행정동명 = 관측소_to_행정동_노원(serial[0])
    관측소 = pd.read_csv(file)
    관측소['행정동명'] = 행정동명
    관측소.drop(['co2', 'vocs', 'noise', 'Unnamed: 10'], axis=1, inplace=True)
    관측소 = 관측소.fillna(0)
    관측소 = 관측소.applymap(lambda x: 0 if x==-999 or x==-9999 else x)
   
    #관측소.replace([-999, -9999], np.nan, inplace=True)
    
    return 관측소
```

```python
def 데이터전처리_종로(file):
    serial = file.split('/')
    serial_name = serial[-1]
    serial = serial_name.split('.')
    행정동명 = 관측소_to_행정동_종로(serial[0])
    관측소 = pd.read_csv(file)
    관측소['행정동명'] = 행정동명
    관측소.drop(['co2', 'vocs', 'noise', 'Unnamed: 10'], axis=1, inplace=True)
    관측소 = 관측소.fillna(0)
    관측소 = 관측소.applymap(lambda x: 0 if x==-999 or x==-9999 else x)
    
    #관측소 = 관측소.apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
    #관측소.replace([-999, -9999], np.nan, inplace=True)
    
    return 관측소
```

#### 따라서 logic에서 concat() 적용 후 datetime() 함수 적용

```python
노원구['tm'] = 노원구.tm.apply(lambda x:str(x))
노원구['tm'] = 노원구.tm.apply(lambda x:datetime.strptime(x,'%Y%m%d%H%M'))
노원구 = 노원구.set_index(['tm'])
```

### 측정시간 Resampling : per 1 hour

```python
노원구_시각화용 = 노원구.pivot_table(index='tm', columns='행정동명',
                aggfunc='median',
                values=['pm10','pm25']
                ).resample('1H').mean()
```
