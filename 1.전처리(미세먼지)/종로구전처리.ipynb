```
# 최빈값 구하는 함수

def mode(gugan_num):
    
    d = {}
    num = list(map(str, gugan_num))
    
    for c in num:
        if c in d:
            d[c] += 1
        else:
            d[c] = 1
    
    return max(d, key=d.get)
```

# 1. 교남동 (단일 스테이션)

```
filepath = '/content/drive/My Drive/수퍼판다/data/air/jongro/{}.csv'.format(종로단일스테이션.iloc[0]['스테이션'])
교남동= pd.read_csv(filepath)
교남동.drop(['Unnamed: 10','co2','vocs', 'flag', 'noise','temp','humi','serial'],axis=1, inplace=True)
교남동['DONG_NM'] = 종로단일스테이션.iloc[0]['행정동']
교남동['tm'] = 교남동.tm.apply(lambda x: str(x))
교남동['tm'] = 교남동.tm.apply(lambda x: datetime.strptime(x, '%Y%m%d%H%M'))
교남동.head()
```

```
# 윗 부분 생략

# 행정동 - nowon3(교남동) - station 1개
# 행정동별 datetime변경 및 이상치 처리 


교남동['tm'] = pd.to_datetime(교남동['tm'], format = '%Y%m%d%H%M')


#하고 싶은 것 :   
#동별 시간대별 미세먼지 농도
# -999, -9999 na로 바꾸고 해보자

교남동['pm10'].replace([-999,-9999], np.NaN, inplace=True)

교남동['pm25'].replace([-999,-9999], np.NaN, inplace=True)   

교남동.describe(include='all')

# 시간데이터 분, 초만 잘라내서 'temp'라는 이름의 변수로 넣어줌
temp = [str(i)[:-6] for i in 교남동.tm]
교남동['temp'] = temp


# 미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 교남동[['temp','pm10']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])

교남동_미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'mise_med'])
교남동_미세먼지['mise_med'].replace([-999,-9999], np.NaN, inplace=True)

# 미세먼지 중앙값 구하기

교남동_median = 교남동_미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='mise_med')

jong1 = pd.DataFrame(교남동_median)


# 미세먼지 최빈값 구하기 
# 위와 같은 일을 하는데 1의자리에서 버린 값을 넣어줌 (최빈구간 구하기 위해)

gugan = []
for i in 교남동[['temp', 'pm10']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_1 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_1.head()



# 전체 데이터에 적용
mise_mode = []
for i in range(len(gugan_1)):
  mise_mode.append(mode(gugan_1.values[i]))


# 초미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 교남동[['temp','pm25']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])


    
교남동_초미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'c_mise_med'])
교남동_초미세먼지['c_mise_med'].replace([-999,-9999], np.NaN, inplace=True)


# 초미세먼지 중앙값 구하기    

교남동_Cmedian = 교남동_초미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='c_mise_med')


jong1_1 = pd.DataFrame(교남동_Cmedian)


# 초미세먼지 최빈값 구하기 

gugan = []
for i in 교남동[['temp', 'pm25']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_3 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_3.head()



# 전체 데이터에 적용
cmise_mode = []
for i in range(len(gugan_3)):
  cmise_mode.append(mode(gugan_3.values[i]))

# 하나의 데이터프레임으로 만들기

jong1.reset_index(inplace=True)
jong1_1.reset_index(inplace=True)

all_jong1 = pd.merge(jong1[['temp', 'mise_med']], jong1_1[['temp','c_mise_med']], on = 'temp')

all_jong1.columns = [['교남동','교남동','교남동'],['temp','mise_med','c_mise_med']]

all_jong1['mise_mode'] = pd.DataFrame(mise_mode)

all_jong1['c_mise_mode'] = pd.DataFrame(cmise_mode)

all_jong1.columns = ['temp','mise_med','c_mise_med','mise_mode','c_mise_mode']

all_jong1['행정동명'] = '교남동'


all_jong1.head()
```

```
# 윗 부분 생략

# 행정동 - nowon3(숭인2동) - station 1개
# 행정동별 datetime변경 및 이상치 처리 


숭인2동['tm'] = pd.to_datetime(숭인2동['tm'], format = '%Y%m%d%H%M')


#하고 싶은 것 :   
#동별 시간대별 미세먼지 농도
# -999, -9999 na로 바꾸고 해보자

숭인2동['pm10'].replace([-999,-9999], np.NaN, inplace=True)

숭인2동['pm25'].replace([-999,-9999], np.NaN, inplace=True)   

숭인2동.describe(include='all')

# 시간데이터 분, 초만 잘라내서 'temp'라는 이름의 변수로 넣어줌
temp = [str(i)[:-6] for i in 숭인2동.tm]
숭인2동['temp'] = temp


# 미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 숭인2동[['temp','pm10']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])

숭인2동_미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'mise_med'])
숭인2동_미세먼지['mise_med'].replace([-999,-9999], np.NaN, inplace=True)

# 미세먼지 중앙값 구하기

숭인2동_median = 숭인2동_미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='mise_med')

jong4 = pd.DataFrame(숭인2동_median)


# 미세먼지 최빈값 구하기 
# 위와 같은 일을 하는데 1의자리에서 버린 값을 넣어줌 (최빈구간 구하기 위해)

gugan = []
for i in 숭인2동[['temp', 'pm10']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_1 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_1.head()



# 전체 데이터에 적용
mise_mode = []
for i in range(len(gugan_1)):
  mise_mode.append(mode(gugan_1.values[i]))


# 초미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 숭인2동[['temp','pm25']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])


    
숭인2동_초미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'c_mise_med'])
숭인2동_초미세먼지['c_mise_med'].replace([-999,-9999], np.NaN, inplace=True)


# 초미세먼지 중앙값 구하기    

숭인2동_Cmedian = 숭인2동_초미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='c_mise_med')


jong4_1 = pd.DataFrame(숭인2동_Cmedian)


# 초미세먼지 최빈값 구하기 

gugan = []
for i in 숭인2동[['temp', 'pm25']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_3 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_3.head()



# 전체 데이터에 적용
cmise_mode = []
for i in range(len(gugan_3)):
  cmise_mode.append(mode(gugan_3.values[i]))

# 하나의 데이터프레임으로 만들기

jong4.reset_index(inplace=True)
jong4_1.reset_index(inplace=True)

all_jong4 = pd.merge(jong4[['temp', 'mise_med']], jong4_1[['temp','c_mise_med']], on = 'temp')

all_jong4.columns = [['숭인2동','숭인2동','숭인2동'],['temp','mise_med','c_mise_med']]

all_jong4['mise_mode'] = pd.DataFrame(mise_mode)

all_jong4['c_mise_mode'] = pd.DataFrame(cmise_mode)

all_jong4.columns = ['temp','mise_med','c_mise_med','mise_mode','c_mise_mode']

all_jong4['행정동명'] = '숭인2동'


all_jong4.head()
```

```
#윗 부분 생략

# 행정동 - nowon3(창신3동) - station 1개
# 행정동별 datetime변경 및 이상치 처리 


창신3동['tm'] = pd.to_datetime(창신3동['tm'], format = '%Y%m%d%H%M')


#하고 싶은 것 :   
#동별 시간대별 미세먼지 농도
# -999, -9999 na로 바꾸고 해보자

창신3동['pm10'].replace([-999,-9999], np.NaN, inplace=True)

창신3동['pm25'].replace([-999,-9999], np.NaN, inplace=True)   

창신3동.describe(include='all')

# 시간데이터 분, 초만 잘라내서 'temp'라는 이름의 변수로 넣어줌
temp = [str(i)[:-6] for i in 창신3동.tm]
창신3동['temp'] = temp


# 미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 창신3동[['temp','pm10']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])

창신3동_미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'mise_med'])
창신3동_미세먼지['mise_med'].replace([-999,-9999], np.NaN, inplace=True)

# 미세먼지 중앙값 구하기

창신3동_median = 창신3동_미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='mise_med')

jong7 = pd.DataFrame(창신3동_median)


# 미세먼지 최빈값 구하기 
# 위와 같은 일을 하는데 1의자리에서 버린 값을 넣어줌 (최빈구간 구하기 위해)

gugan = []
for i in 창신3동[['temp', 'pm10']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_1 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_1.head()



# 전체 데이터에 적용
mise_mode = []
for i in range(len(gugan_1)):
  mise_mode.append(mode(gugan_1.values[i]))


# 초미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 창신3동[['temp','pm25']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])


    
창신3동_초미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'c_mise_med'])
창신3동_초미세먼지['c_mise_med'].replace([-999,-9999], np.NaN, inplace=True)


# 초미세먼지 중앙값 구하기    

창신3동_Cmedian = 창신3동_초미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='c_mise_med')


jong7_1 = pd.DataFrame(창신3동_Cmedian)


# 초미세먼지 최빈값 구하기 

gugan = []
for i in 창신3동[['temp', 'pm25']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_3 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_3.head()



# 전체 데이터에 적용
cmise_mode = []
for i in range(len(gugan_3)):
  cmise_mode.append(mode(gugan_3.values[i]))

# 하나의 데이터프레임으로 만들기

jong7.reset_index(inplace=True)
jong7_1.reset_index(inplace=True)

all_jong7 = pd.merge(jong7[['temp', 'mise_med']], jong7_1[['temp','c_mise_med']], on = 'temp')

all_jong7.columns = [['창신3동','창신3동','창신3동'],['temp','mise_med','c_mise_med']]

all_jong7['mise_mode'] = pd.DataFrame(mise_mode)

all_jong7['c_mise_mode'] = pd.DataFrame(cmise_mode)

all_jong7.columns = ['temp','mise_med','c_mise_med','mise_mode','c_mise_mode']

all_jong7['행정동명'] = '창신3동'


all_jong7.head()
```

```
# 행정동 - nowon3(평창동) - station 1개
# 행정동별 datetime변경 및 이상치 처리 


평창동['tm'] = pd.to_datetime(평창동['tm'], format = '%Y%m%d%H%M')


#하고 싶은 것 :   
#동별 시간대별 미세먼지 농도
# -999, -9999 na로 바꾸고 해보자

평창동['pm10'].replace([-999,-9999], np.NaN, inplace=True)

평창동['pm25'].replace([-999,-9999], np.NaN, inplace=True)   

평창동.describe(include='all')

# 시간데이터 분, 초만 잘라내서 'temp'라는 이름의 변수로 넣어줌
temp = [str(i)[:-6] for i in 평창동.tm]
평창동['temp'] = temp


# 미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 평창동[['temp','pm10']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])

평창동_미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'mise_med'])
평창동_미세먼지['mise_med'].replace([-999,-9999], np.NaN, inplace=True)

# 미세먼지 중앙값 구하기

평창동_median = 평창동_미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='mise_med')

jong10 = pd.DataFrame(평창동_median)


# 미세먼지 최빈값 구하기 
# 위와 같은 일을 하는데 1의자리에서 버린 값을 넣어줌 (최빈구간 구하기 위해)

gugan = []
for i in 평창동[['temp', 'pm10']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_1 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_1.head()



# 전체 데이터에 적용
mise_mode = []
for i in range(len(gugan_1)):
  mise_mode.append(mode(gugan_1.values[i]))


# 초미세먼지
# 같은 동 내 다른 스테이션에서 측정된 미세먼지농도를 하나의 리스트로 묶은 것 'mise'라는 이름의 변수를 가지는 데이터프레임 생성

tlist = []
for i in 평창동[['temp','pm25']].values:   # 미세먼지 데이터에서 시간(에서 분,초 뺀 것), 미세먼지만 떼어 냄
    tlist.append([i[0],i[1]])


    
평창동_초미세먼지 = pd.DataFrame(tlist, columns = ['temp', 'c_mise_med'])
평창동_초미세먼지['c_mise_med'].replace([-999,-9999], np.NaN, inplace=True)


# 초미세먼지 중앙값 구하기    

평창동_Cmedian = 평창동_초미세먼지.pivot_table(index='temp',
                           aggfunc='median', 
                           values='c_mise_med')


jong10_1 = pd.DataFrame(평창동_Cmedian)


# 초미세먼지 최빈값 구하기 

gugan = []
for i in 평창동[['temp', 'pm25']].values:
     gugan.append([i[0],[(i[1]//10)*10]])


gugan_3 = pd.DataFrame(gugan, columns = ['temp', 'gugan']).groupby('temp')['gugan'].sum()
gugan_3.head()



# 전체 데이터에 적용
cmise_mode = []
for i in range(len(gugan_3)):
  cmise_mode.append(mode(gugan_3.values[i]))

# 하나의 데이터프레임으로 만들기

jong10.reset_index(inplace=True)
jong10_1.reset_index(inplace=True)

all_jong10 = pd.merge(jong10[['temp', 'mise_med']], jong10_1[['temp','c_mise_med']], on = 'temp')

all_jong10.columns = [['평창동','평창동','평창동'],['temp','mise_med','c_mise_med']]

all_jong10['mise_mode'] = pd.DataFrame(mise_mode)

all_jong10['c_mise_mode'] = pd.DataFrame(cmise_mode)

all_jong10.columns = ['temp','mise_med','c_mise_med','mise_mode','c_mise_mode']

all_jong10['행정동명'] = '평창동'


all_jong10.head()
```
