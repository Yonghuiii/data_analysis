# 1. V10O1610102(하계1동)

## 2018년 5월 7일부터 데이터가 있음, 그전까지의 데이터는 없음

## 2018년 9월 16일 데이터 NaN, 그리고 18일부터 데이터 X

## 2018년 10월 1~9일까지 데이터 X

```python
df1= pd.read_csv('환경기상데이터/노원구/V10O1610102.csv')
df1.drop(['co2', 'vocs', 'noise', 'Unnamed: 10'], axis=1, inplace=True)
df1 = df1.applymap(lambda x: 0 if x==-999 or x==-9999 else x)
df1['tm'] = df1.tm.apply(lambda x: str(x))
df1['tm'] = df1.tm.apply(lambda x: datetime.strptime(x, '%Y%m%d%H%M'))
df1.set_index('tm', inplace = True)
df1.info()

for i in range(5, 13, 1):
    num = df1['2018-{0}'.format(i)].shape
    if i ==6 or i==9 or i==11 :
        print('{0}월:'.format(i), num[0],
                  ', 원래값과 차이', 43200 - num[0])
    else:
        print('{0}월:'.format(i), num[0],
              ', 원래값과 차이', 44640-num[0])
        
for i in range(1, 4, 1):
    num = df1['2019-{0}'.format(i)].shape
    if i == 2:
        print('{0}월:'.format(i), num[0],
              ', 원래값과 차이', 40320 - num[0])
    else:
        print('{0}월:'.format(i), num[0],
              ', 원래값과 차이',  44640 - num[0])
```

```python
df1['2018-10'].resample('D').mean()
```

```python
# pm25가 NaN인 시간들 찾아내기

df1['pm25'].isnull().index
```
