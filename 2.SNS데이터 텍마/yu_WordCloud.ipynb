```
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```
def displayWordCloud(data=None, backgroundcolor='white', width=1600, height=800):
    wordcloud = WordCloud(
                        font_path='/content/drive/My Drive/수퍼판다/SNS데이터 텍마/NANUMGOTHIC.TTF',
                        stopwords=STOPWORDS,
                        background_color=backgroundcolor,
                        width=width, height=height).generate(data)
    plt.figure(figsize=(15, 10))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```
from konlpy.tag import Kkma
kkma = Kkma()
```

```
def kkma_nouns_wordcloud(content):
  kkma_content_nouns = kkma.nouns(''.join(str(content.fillna(''))))
  displayWordCloud(' '.join(kkma_content_nouns))
```

```
# 뉴스, TITLE
STOPWORDS = ['대한', '민국', '대한민국', '만들기']
kkma_nouns_wordcloud(news['TITLE'])
```

```
# 뉴스, CONTENT
STOPWORDS = ['취임', '앵커', '김재현', '재현', '청장', '산림청', '산림청청장']
kkma_nouns_wordcloud(news['CONTENT'])
```
